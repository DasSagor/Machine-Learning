```python
import numpy as pd
import pandas as pd

data = [120,125,128,122,126,124,600]

df = pd.DataFrame({'HousePrice': data})

mean_value = df['HousePrice'].mean()

print(mean_value)
```

```python
median_value = df['HousePrice'].median()
print(median_value)
```

```python
df['City'] = ['Dhaka','Dhaka', 'Chattogram','Sylet','Dhaka','Rajshahi','Chattogram']

mode_value = df['City'].mode()[0]

print(mode_value)
```
